```python
import requests
from bs4 import BeautifulSoup as bs
import urllib.parse
import pandas as pd

job_name=input('Enter job name: ')
parse_job_name=urllib.parse.quote(job_name)

job_country=input('Enter job country: ')

url=f'https://www.linkedin.com/jobs/search?keywords={parse_job_name}&location=egypt&geoId=106155005&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0'
Job_titles, companies,companies_link, locations, links, Experience_Needed,publish_time =[], [], [], [], [], [], []

response=requests.get(url)
soup=bs(response.content, 'html.parser')
soup
```

```python
containers = soup.find_all("div", class_="base-card relative w-full hover:no-underline focus:no-underline base-card--link base-search-card base-search-card--link job-search-card")
len(containers)
```

```python
for container in containers:
  job_title = container.find("h3", class_="base-search-card__title")
  Job_titles.append(job_title.text.strip())

  company = container.find("h4", class_="base-search-card__subtitle")
  companies.append(company.text.strip())

  company_link=container.find("a", class_="hidden-nested-link")
  companies_link.append(company_link.get('href'))

  location = container.find("span", class_="job-search-card__location")
  locations.append(location.text.strip())

  time = container.find("time")
  print(time.text.strip())
  publish_time.append(time.text.strip())


  link = container.find("a", class_="base-card__full-link")
  links.append(link.get('href'))
```

```python
df=pd.DataFrame({'Job Title':Job_titles,'Company':companies,'company link':companies_link,'Location':locations,'Publish Time':publish_time,'link':links})
df
```

```python
df.to_csv(f'jobs_{job_name}.csv', index=False)
```
